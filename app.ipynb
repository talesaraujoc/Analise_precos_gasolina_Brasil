### DESAFIO: Crie uma tabela contendo uma serie temporal (mensal) com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados.

```python
df_gasolina_comum
```

```python
df_diff = pd.DataFrame()
```

```python
df_max = df_gasolina_comum.groupby('ANO-MÊS').max()['PREÇO MÉDIO REVENDA']

df_min = df_gasolina_comum.groupby('ANO-MÊS').min()['PREÇO MÉDIO REVENDA']
```

```python
idxmax = df_gasolina_comum.groupby('ANO-MÊS')['PREÇO MÉDIO REVENDA'].idxmax()
idxmin = df_gasolina_comum.groupby('ANO-MÊS')['PREÇO MÉDIO REVENDA'].idxmin()
```

```python
df_diff['Dif_Absoluta'] = df_max - df_min
df_diff['Dif_Perc'] = (df_max -df_min) / df_min * 100
df_diff['Máx'] = df_max
df_diff['Mín'] = df_min
```

```python
df_diff
```

```python
df_diff['Estado Máx'] = df_gasolina_comum.loc[idxmax, :]['ESTADO'].values
```

```python
df_diff['Estado Mín'] = df_gasolina_comum.loc[idxmin, :]['ESTADO'].values
```
